```python
import pandas as pd
```

```python
df = pd.read_csv("data/moj_pqs.csv")
df.head()
```

```python
df = df[["body", "answer"]]
```

```python
import json

my_json = df.loc[2, "answer"]
j = json.loads(my_json)
j
```

```python
def parse_to_answer(x):
    j = json.loads(x)
    return "\n ".join([i["answer"] for i in j])
```

```python
df["parsed_answer"] = df["answer"].apply(parse_to_answer)
```

```python
final_df = df[["body", "parsed_answer"]]
```

```python
final_df.to_csv("all_question_utf_8.csv")
```
